## **Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.3, stratify= y, random_state=2529)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

## **Get Model Train**

```python
from sklearn.linear_model import LogisticRegression
```

```python
hill = LogisticRegression(max_iter=10000)
```

```python
hill.fit(x_train,y_train)
```

## **Get Model Prediction**

```python
y_pred = hill.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

## **Get Probability of Each Predicted Class**

```python
hill.predict_proba(x_test)
```

## **Get Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## **Explaination**

Logistic regression predicts binary or categorical outcomes by estimating the relationship between input variables and the log-odds of the outcome. By applying the logistic function, it converts the log-odds to probabilities, allowing for predictions to be made.
